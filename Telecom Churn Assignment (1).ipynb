```python
#Interquartile Range
Q1=dftelecom.quantile(0.25)
Q2=dftelecom.quantile(0.5)
Q3=dftelecom.quantile(0.75)
IQR=Q3-Q1
print("Q1\n",Q1,"\nQ2\n",Q2,"\nQ3\n",Q3,"\nIQR\n",IQR)
```

```python
sns.boxplot(dftelecom['Total day minutes']) ## Finding outliers for Total Day Minutes
```

```python
Q1totalday=dftelecom['Total day minutes'].quantile(0.25)
Q2totalday=dftelecom['Total day minutes'].quantile(0.5)
Q3totalday=dftelecom['Total day minutes'].quantile(0.75)
IQRtotalday=Q3totalday-Q1totalday

# Range is Q1-1.5*IQR < value < Q3+1.5*IQR

floortotalday=Q1totalday-1.5*IQRtotalday
captotalday=Q3totalday+1.5*IQRtotalday
print("Q1 ----- ",Q1totalday,"\nQ2 -----",Q2totalday,"\nQ3 ------ ",Q3totalday,"\nIQR ------",IQRtotalday)
print("Floor ----- ",floortotalday,"\nCap ------ ",captotalday)
```

```python
# for Lower LImit : 5th percentile 
# For Upper Limit : 95th Percentile

Q5totalday=dftelecom['Total day minutes'].quantile(0.05)
Q95totalday=dftelecom['Total day minutes'].quantile(0.95)
print("Q5 is ",Q5totalday,"\nQ95 is ",Q95totalday)
```

```python
dftelecom.describe()
```

```python
# Setting the lower limit Q5 as 90 and Upper limit Q95 as 270:

dftelecom.loc[dftelecom['Total day minutes']>270,'Total day minutes']=270
dftelecom.loc[dftelecom['Total day minutes']<90,'Total day minutes']=90
dftelecom['Total day minutes'].describe()
```

```python
Q1totalnight=dftelecom['Total night minutes'].quantile(0.25)
Q2totalnight=dftelecom['Total night minutes'].quantile(0.5)
Q3totalnight=dftelecom['Total night minutes'].quantile(0.75)
IQRtotalnight=Q3totalnight-Q1totalnight

# Range -- Q1-1.5*IQR<value<Q3+1.5*IQR

floortotalnight=Q1totalnight-1.5*IQRtotalnight
captotalnight=Q3totalnight+1.5*IQRtotalnight
print("Q1 is ",Q1totalnight,"\nQ2 is ",Q2totalnight,"\nQ3 is ",Q3totalnight,"\nIQR is",IQRtotalnight)
print("Floor is ",floortotalnight,"\nCap is ",captotalnight)
```

```python
# for floor we can use 5th percentile 
# For Cap we can use 95th Percentile

Q5totalnight=dftelecom['Total night minutes'].quantile(0.05)
Q95totalnight=dftelecom['Total night minutes'].quantile(0.95)
print("Q5 is",Q5totalnight,"\nQ95 is ",Q95totalnight)
```

```python
dftelecom['Total night minutes'].describe()
```

```python
# Setting the lower and upper limit for Total Night Minutes to Q5 and Q95 i.e 118 and 283

dftelecom.loc[dftelecom['Total night minutes']>283,'Total night minutes']=283
dftelecom.loc[dftelecom['Total night minutes']<118,'Total night minutes']=118
dftelecom['Total night minutes'].describe()
```
